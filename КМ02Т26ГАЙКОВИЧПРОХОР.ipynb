## 1. Иерархия встроенных типов в Python

### Задание 1
Построение и визуализация иерархической модели встроенных типов объектов Python.

Ниже представлена иерархическая модель встроенных типов Python с указанием:
- примера литерала для создания объекта
- упорядоченности элементов (для коллекций)
- изменяемости/неизменяемости объектов

```
                        Встроенные типы Python
                                │
        ┌───────────────────────┼───────────────────────┐
        │                       │                       │
    Числовые               Коллекции                 Прочие
        │                       │                       │
   ┌────┼────┐         ┌────────┼────────┐         ┌────┼────┐
   │    │    │         │        │        │         │    │    │
  int float complex  Последов. Отображ. Множеств. bool None type
  42  3.14 2+3j       │      dict     │        True None type(x)
  неизм неизм неизм   │    {'a':1}    │        неизм
                       │    изменяем  set frozenset
               ┌───────┼──────┐   {1,2}  frozenset({1})
               │       │      │   изменяем неизменяем
              str    list   tuple
             'abc'  [1,2]  (1,2)
           упоряд  упоряд  упоряд
           неизм   изменяем неизм
```

**Пояснения:**
- **Упорядоченные коллекции** (сохраняют порядок): `str`, `list`, `tuple`; `dict` (с Python 3.7+ сохраняет порядок вставки)
- **Неупорядоченные коллекции**: `set`, `frozenset`
- **Изменяемые (mutable)**: `list`, `dict`, `set`
- **Неизменяемые (immutable)**: `int`, `float`, `complex`, `str`, `tuple`, `frozenset`, `bool`, `NoneType`

#### Раздел 2.3. Встроенные типы последовательностей: str, list, tuple

**Задача:** Определить строку `S = 'spam'`. Написать выражение, которое изменяет строку на `'slam'`, используя нарезание и конкатенацию. Рассмотреть также индексацию, присваивание по индексу. Привести аналогичные примеры для списка и кортежа.


````python
# Исходная строка
S = 'spam'
print(f"Исходная строка: {S}")

# Способ 1: нарезание (slicing) + конкатенация
S_new = S[0] + 'l' + S[2:]
print(f"Нарезание + конкатенация: {S_new}")

# Способ 2: индексация + конкатенация
S_new2 = S[0] + 'l' + S[2] + S[3]
print(f"Индексация + конкатенация: {S_new2}")

# Способ 3: попытка присваивания по индексу (ОШИБКА — строки неизменяемы)
try:
    S[1] = 'l'
except TypeError as e:
    print(f"Присваивание по индексу для строки: ошибка — {e}")
````

````python
# Аналогичные операции для списка
L = ['s', 'p', 'a', 'm']
print(f"Исходный список: {L}")

# Список — изменяемый объект, поэтому можно присваивать по индексу
L[1] = 'l'
print(f"Присваивание по индексу: {L}")

# Через нарезание + конкатенацию (создаёт НОВЫЙ список)
L2 = ['s', 'p', 'a', 'm']
L2_new = L2[:1] + ['l'] + L2[2:]
print(f"Нарезание + конкатенация: {L2_new}")
print(f"Объекты разные? id(L2)={id(L2)}, id(L2_new)={id(L2_new)}")
````

````python
# Аналогичные операции для кортежа
T = ('s', 'p', 'a', 'm')
print(f"Исходный кортеж: {T}")

# Кортеж — неизменяемый, как и строка. Используем нарезание + конкатенацию
T_new = T[:1] + ('l',) + T[2:]
print(f"Нарезание + конкатенация: {T_new}")

# Попытка присваивания по индексу — ошибка
try:
    T[1] = 'l'
except TypeError as e:
    print(f"Присваивание по индексу для кортежа: ошибка — {e}")

# Вывод: и строки, и кортежи неизменяемы — для модификации нужно создавать новый объект
````

**Задача:** Удаление повторяющихся символов строки с использованием множества.


````python
# Удаление повторяющихся символов строки
original = 'абракадабра'
print(f"Исходная строка: '{original}'")

# Преобразуем строку в множество — автоматически удаляются дубликаты
unique_chars = set(original)
print(f"Уникальные символы (множество): {unique_chars}")

# Если нужно получить строку обратно (порядок может измениться!)
unique_str = ''.join(sorted(unique_chars))
print(f"Строка из уникальных символов (отсортированная): '{unique_str}'")

# Ещё пример — английский текст
text = 'mississippi'
print(f"\n'{text}' -> уникальные символы: {set(text)}")
print(f"Количество уникальных символов: {len(set(text))}")
````

#### Раздел 2.8. Интерфейс объекта: операции и методы

**Задача 1:** Соединение списка строк в одну строку (join) и разбиение строки на список подстрок (split).


````python
# Метод str.join(iterable) — соединяет элементы iterable через строку-разделитель
# Спецификация: S.join(iterable) -> str
# Возвращает строку, полученную конкатенацией элементов iterable через S

words = ['Python', 'это', 'мощный', 'язык']
separator = ' '
result = separator.join(words)
print(f"Список: {words}")
print(f"Результат join(' '): '{result}'")

# Соединение через дефис
result2 = '-'.join(['2026', '03', '13'])
print(f"Через дефис: '{result2}'")
````

````python
# Метод str.split(sep=None, maxsplit=-1) -> list[str]
# Спецификация: разбивает строку по разделителю sep, возвращает список подстрок
# Если sep=None, разбиение по пробельным символам

sentence = 'Python это мощный язык'
parts = sentence.split(' ')
print(f"Строка: '{sentence}'")
print(f"Результат split(' '): {parts}")

# Разбиение по запятой
csv_line = 'Минск,Гомель,Брест,Витебск'
cities = csv_line.split(',')
print(f"\nCSV: '{csv_line}'")
print(f"Результат split(','): {cities}")
````

**Задача 2:** Три способа добавления элемента в список.

````python
# Способ 1: метод append — добавляет один элемент в конец
lst = [1, 2, 3]
lst.append(4)
print(f"append(4): {lst}")

# Способ 2: метод insert — вставляет элемент по заданному индексу
lst2 = [1, 2, 3]
lst2.insert(1, 99)  # вставить 99 на позицию 1
print(f"insert(1, 99): {lst2}")

# Способ 3: метод extend — добавляет все элементы итерируемого объекта
lst3 = [1, 2, 3]
lst3.extend([4, 5])
print(f"extend([4, 5]): {lst3}")

# Бонус: через конкатенацию (создаёт новый объект, а не модифицирует)
lst4 = [1, 2, 3]
lst4 = lst4 + [4]
print(f"Конкатенация + [4]: {lst4}")
````

**Задача 3:** Перестановка элементов в обратном порядке (операция и метод). Аналогия для строки и кортежа.

````python
# Список: метод reverse() — модифицирует на месте (in-place)
lst = [10, 20, 30, 40, 50]
print(f"Исходный список: {lst}")
lst.reverse()
print(f"После reverse(): {lst}")

# Список: операция нарезания [::-1] — создаёт новый перевёрнутый список
lst2 = [10, 20, 30, 40, 50]
lst2_rev = lst2[::-1]
print(f"Через [::-1]: {lst2_rev} (оригинал: {lst2})")
````

````python
# Строка — неизменяемая, метод reverse() отсутствует
s = 'Привет'
# s.reverse()  # AttributeError — у строки нет метода reverse
s_rev = s[::-1]
print(f"Строка '{s}' в обратном порядке: '{s_rev}'")

# Кортеж — неизменяемый, метод reverse() отсутствует
t = (1, 2, 3, 4, 5)
# t.reverse()  # AttributeError — у кортежа нет метода reverse
t_rev = t[::-1]
print(f"Кортеж {t} в обратном порядке: {t_rev}")

# Вывод: для неизменяемых типов (str, tuple) используется только нарезание [::-1],
# которое создаёт новый объект. Метод reverse() доступен только для списков.
````

**Задача 4:** Сортировка списка с аргументом key.

````python
# Сортировка с параметром key
# Метод list.sort(key=None, reverse=False) — сортирует список на месте

words = ['банан', 'Яблоко', 'вишня', 'Арбуз', 'груша']
print(f"До сортировки: {words}")

# Сортировка по умолчанию — с учётом регистра (заглавные буквы идут первыми)
words_copy = words.copy()
words_copy.sort()
print(f"sort() по умолчанию: {words_copy}")

# Сортировка без учёта регистра — key=str.lower
words.sort(key=str.lower)
print(f"sort(key=str.lower): {words}")

# Сортировка по длине слова
data = ['кот', 'медведь', 'лев', 'жираф', 'волк']
data.sort(key=len)
print(f"\nСортировка по длине: {data}")

# Комментарий: параметр key задаёт функцию, результат которой используется
# для сравнения элементов. Сами элементы при этом не изменяются.
````

**Задача 2:** Объект с четырьмя ссылками. Подсчёт количества ссылок.

````python
import sys

# Создаём объект и четыре ссылки на него
original = [100, 200, 300]
ref_a = original
ref_b = original
ref_c = original

# Все переменные ссылаются на один и тот же объект
print(f"id(original) = {id(original)}")
print(f"id(ref_a)    = {id(ref_a)}")
print(f"id(ref_b)    = {id(ref_b)}")
print(f"id(ref_c)    = {id(ref_c)}")

# Количество ссылок на объект
# sys.getrefcount() добавляет +1 за саму передачу аргумента
ref_count = sys.getrefcount(original)
print(f"\nsys.getrefcount(original) = {ref_count}")
print(f"Реальное количество ссылок ≈ {ref_count} (включая временную ссылку getrefcount)")
````

**Задача 3:** Четыре способа копирования элементов списка. Отличие copy от deepcopy.

````python
import copy

# Исходный список (содержит вложенный список)
original = [[1, 2], [3, 4], 'текст']

# Способ 1: метод copy()
copy1 = original.copy()

# Способ 2: нарезание [:]
copy2 = original[:]

# Способ 3: конструктор list()
copy3 = list(original)

# Способ 4: copy.copy() (поверхностная копия)
copy4 = copy.copy(original)

print("Все 4 способа создают поверхностные копии:")
print(f"original:      {original}")
print(f"copy1 (.copy): {copy1}")
print(f"copy2 ([:]   ): {copy2}")
print(f"copy3 (list()): {copy3}")
print(f"copy4 (copy.copy): {copy4}")
````

````python
import copy

# Демонстрация отличия поверхностной копии (copy) от глубокой (deepcopy)
original = [[1, 2, 3], [4, 5, 6]]

shallow = original.copy()       # поверхностная копия
deep = copy.deepcopy(original)  # глубокая копия

# Модифицируем вложенный список оригинала
original[0][0] = 999

print(f"original после изменения: {original}")
print(f"shallow (поверхностная):  {shallow}")   # ИЗМЕНИЛСЯ! [0][0] == 999
print(f"deep (глубокая):          {deep}")       # НЕ изменился! [0][0] == 1

# Пояснение: поверхностная копия (copy) создаёт новый внешний список,
# но вложенные объекты НЕ копируются — они разделяются между оригиналом и копией.
# Глубокая копия (deepcopy) рекурсивно копирует ВСЕ вложенные объекты.
print(f"\nshallow[0] is original[0]: {shallow[0] is original[0]}")  # True
print(f"deep[0] is original[0]:    {deep[0] is original[0]}")      # False
````

#### Раздел 2.10. Включения (comprehensions)

**Задача 1:** Создание множества с использованием включения с условием.


````python
# Множество чётных квадратов чисел от 1 до 20
even_squares = {x**2 for x in range(1, 21) if x % 2 == 0}
print(f"Чётные квадраты (1..20): {even_squares}")

# Множество гласных букв в тексте
text = 'Компьютерная математика и системный анализ'
vowels = {'а', 'е', 'ё', 'и', 'о', 'у', 'ы', 'э', 'ю', 'я'}
found_vowels = {ch.lower() for ch in text if ch.lower() in vowels}
print(f"Гласные в тексте: {found_vowels}")
````

**Задача 2:** Числовая матрица 7×8 через вложенные включения и через numpy.

````python
# Матрица 7x8 через вложенные включения: элемент = i * 8 + j + 1
matrix = [[i * 8 + j + 1 for j in range(8)] for i in range(7)]
print("Матрица 7×8 (list comprehension):")
for row in matrix:
    print(row)
````

````python
import numpy as np

# Аналогичная матрица через numpy
matrix_np = np.arange(1, 57).reshape(7, 8)
print("Матрица 7×8 (numpy):")
print(matrix_np)
print(f"\nТип: {type(matrix_np)}, размерность: {matrix_np.shape}")
````

**Задача 3:** Словарь с включением и двумя уровнями вложенности.

````python
# Словарь: для каждой буквы — вложенный словарь с кодами символов
# Пример с двумя уровнями вложенности
alphabet_info = {
    letter: {f'код_{enc}': letter.encode(enc) for enc in ['utf-8', 'ascii']}
    for letter in 'ABCDEF'
}
print("Словарь с двумя уровнями вложенности:")
for key, val in alphabet_info.items():
    print(f"  {key}: {val}")

# Ещё пример: таблица умножения как словарь словарей
mult_table = {
    i: {j: i * j for j in range(1, 4)}
    for i in range(1, 4)
}
print(f"\nТаблица умножения: {mult_table}")
print(f"3 × 2 = {mult_table[3][2]}")
````

#### Раздел 2.11. Форматированный вывод строк

**Задача 1:** Форматирование переменной через f-string `{variable=:…}`, оператор `%` и метод `format`.


````python
# Переменная для форматирования
pi = 3.141592653589793

# Способ 1: f-строка с синтаксисом variable=
print(f"{pi=:.4f}")
# Выводит: pi=3.1416 — отладочный формат, показывает имя переменной и значение

# Способ 2: бинарная операция %
print("pi=%.4f" % pi)

# Способ 3: метод format
print("pi={:.4f}".format(pi))
````

**Задача 2:** Форматирование выражения через f-string `{expression=:…}`, оператор `%` и метод `format`.

````python
x = 7
y = 3

# Способ 1: f-строка с выражением
print(f"{x + y=:>10d}")
# Показывает выражение и его результат с выравниванием вправо в поле шириной 10

# Способ 2: бинарная операция %
print("x + y=%10d" % (x + y))

# Способ 3: метод format
print("x + y={:>10d}".format(x + y))
````

#### Раздел 2.12. Файловый объект

**Задача:** Примеры использования функций и методов: `open`, `close`, `read`, `readline`, `readlines`, `write`, `writelines`.


````python
# Запись в файл: open, write, writelines, close
f = open('example.txt', 'w', encoding='utf-8')
f.write('Первая строка файла\n')            # write — записывает строку
f.write('Вторая строка файла\n')
f.writelines(['Третья строка\n',            # writelines — записывает список строк
              'Четвёртая строка\n',
              'Пятая строка\n'])
f.close()                                    # close — закрывает файл
print("Файл записан успешно.")
````

````python
# Чтение файла целиком: read
f = open('example.txt', 'r', encoding='utf-8')
content = f.read()           # read — читает весь файл как одну строку
f.close()
print("read() — весь файл:")
print(content)
````

````python
# Чтение построчно: readline
f = open('example.txt', 'r', encoding='utf-8')
line1 = f.readline()         # readline — читает одну строку
line2 = f.readline()
f.close()
print(f"readline() 1: {line1!r}")
print(f"readline() 2: {line2!r}")
````

````python
# Чтение всех строк в список: readlines
f = open('example.txt', 'r', encoding='utf-8')
lines = f.readlines()        # readlines — возвращает список всех строк
f.close()
print(f"readlines(): {lines}")
print(f"Количество строк: {len(lines)}")

# Рекомендуемый способ — менеджер контекста (автоматически закрывает файл)
with open('example.txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f, 1):
        print(f"Строка {i}: {line.strip()}")
````

### Задание 2.3 — Фрагменты кода с ошибками

Ниже приведены типичные ошибки при работе со встроенными типами Python (на основе лекционного материала), их причины и исправления.


**Ошибка 1:** Попытка изменить символ строки по индексу (строки неизменяемы)

````python
# Ошибочный код:
s = 'hello'
try:
    s[0] = 'H'
except TypeError as e:
    print(f"Ошибка: {e}")
    # Причина: строка — неизменяемый (immutable) объект

# Исправление: создать новую строку
s = 'H' + s[1:]
print(f"Исправлено: {s}")
````

**Ошибка 2:** Обращение к несуществующему ключу словаря

````python
# Ошибочный код:
d = {'a': 1, 'b': 2}
try:
    val = d['c']
except KeyError as e:
    print(f"Ошибка: KeyError — ключ {e} не найден в словаре")
    # Причина: ключ 'c' отсутствует в словаре

# Исправление: использовать метод get с значением по умолчанию
val = d.get('c', 0)
print(f"Исправлено (get): {val}")
````

**Ошибка 3:** Попытка сложить строку и число

````python
# Ошибочный код:
try:
    result = 'возраст: ' + 21
except TypeError as e:
    print(f"Ошибка: {e}")
    # Причина: Python не выполняет неявное преобразование int к str

# Исправление: явное преобразование
result = 'возраст: ' + str(21)
print(f"Исправлено: {result}")
````

**Ошибка 4:** Выход за пределы индекса списка

````python
# Ошибочный код:
lst = [10, 20, 30]
try:
    print(lst[5])
except IndexError as e:
    print(f"Ошибка: {e}")
    # Причина: индекс 5 выходит за пределы списка длиной 3

# Исправление: проверить длину или использовать допустимый индекс
print(f"Исправлено: lst[-1] = {lst[-1]}")
````

**Ошибка 5:** Добавление изменяемого элемента (списка) в множество

````python
# Ошибочный код:
try:
    s = {1, 2, [3, 4]}
except TypeError as e:
    print(f"Ошибка: {e}")
    # Причина: список — нехешируемый (unhashable) объект,
    # а множество может содержать только хешируемые элементы

# Исправление: использовать кортеж вместо списка
s = {1, 2, (3, 4)}
print(f"Исправлено (кортеж): {s}")
````

**Ошибка 6:** Модификация словаря при итерации

````python
# Ошибочный код:
d = {'a': 1, 'b': 2, 'c': 3}
try:
    for key in d:
        if d[key] < 2:
            del d[key]
except RuntimeError as e:
    print(f"Ошибка: {e}")
    # Причина: нельзя изменять размер словаря во время итерации по нему

# Исправление: итерировать по копии ключей
d = {'a': 1, 'b': 2, 'c': 3}
for key in list(d.keys()):
    if d[key] < 2:
        del d[key]
print(f"Исправлено: {d}")
````

**Ошибка 7:** Попытка вызвать метод sort у кортежа

````python
# Ошибочный код:
t = (5, 2, 8, 1)
try:
    t.sort()
except AttributeError as e:
    print(f"Ошибка: {e}")
    # Причина: кортеж — неизменяемый, метод sort() отсутствует

# Исправление: использовать встроенную функцию sorted()
t_sorted = tuple(sorted(t))
print(f"Исправлено: {t_sorted}")
````

**Ошибка 8:** Деление на ноль

````python
# Ошибочный код:
try:
    x = 10 / 0
except ZeroDivisionError as e:
    print(f"Ошибка: {e}")
    # Причина: деление на ноль математически не определено

# Исправление: проверка делителя
divisor = 0
x = 10 / divisor if divisor != 0 else float('inf')
print(f"Исправлено: {x}")
````

**Ошибка 9:** Использование изменяемого объекта как значения по умолчанию

````python
# Ошибочный код:
def add_item_bad(item, lst=[]):
    lst.append(item)
    return lst

# При повторных вызовах список «накапливается» — неожиданное поведение
print(add_item_bad(1))  # [1]
print(add_item_bad(2))  # [1, 2] — ожидалось [2]!
# Причина: аргумент по умолчанию создаётся один раз при определении функции

# Исправление: использовать None как значение по умолчанию
def add_item_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_good(1))  # [1]
print(add_item_good(2))  # [2] — правильно
````

### Задание 2.4 — Рекомендация из PEP 8

**Рекомендация:** *Используйте 4 пробела для каждого уровня отступа.* Не используйте символы табуляции для создания отступов. Пробелы являются предпочтительным способом создания отступов в Python.

Эта рекомендация критически важна, поскольку в Python отступы определяют структуру блоков кода (в отличие от фигурных скобок в C/C++/Java).


````python
# Пример кода, соответствующего PEP 8 (4 пробела на уровень):

def calculate_statistics(numbers):
    """Вычисляет базовые статистики для списка чисел."""
    if not numbers:
        return None

    total = sum(numbers)
    count = len(numbers)
    mean = total / count

    sorted_nums = sorted(numbers)
    mid = count // 2
    if count % 2 == 0:
        median = (sorted_nums[mid - 1] + sorted_nums[mid]) / 2
    else:
        median = sorted_nums[mid]

    return {'среднее': mean, 'медиана': median, 'сумма': total}


data = [15, 8, 23, 42, 4, 16]
stats = calculate_statistics(data)
print(f"Данные: {data}")
print(f"Статистика: {stats}")

# Каждый уровень вложенности (def, if, else) добавляет ровно 4 пробела
````

### Глава 5. Числовые типы

**Вопрос:** Чем отличается целочисленное деление `//` от обычного деления `/` в Python 3?

**Ответ:** Оператор `/` всегда возвращает результат типа `float`, даже если оба операнда целые. Оператор `//` выполняет деление с округлением вниз (к отрицательной бесконечности) и возвращает целое число, если оба операнда целые.


````python
# Разница между / и //
print(f"7 / 2 = {7 / 2}")       # 3.5 (float)
print(f"7 // 2 = {7 // 2}")     # 3 (int, округление вниз)
print(f"-7 // 2 = {-7 // 2}")   # -4 (округление к -∞, а не к нулю!)
print(f"7.0 // 2 = {7.0 // 2}") # 3.0 (float, т.к. один из операндов — float)

# divmod возвращает пару (частное, остаток)
print(f"divmod(7, 2) = {divmod(7, 2)}")
````

### Глава 6. Строки

**Вопрос:** Что такое нарезание (slicing) строк и как оно работает?

**Ответ:** Нарезание — это извлечение подстроки с использованием синтаксиса `S[start:end:step]`, где `start` — начальный индекс (включительно), `end` — конечный (исключительно), `step` — шаг. Все параметры необязательны, по умолчанию: `start=0`, `end=len(S)`, `step=1`.


````python
S = 'Лабораторная работа'
print(f"Строка: '{S}'")
print(f"S[0:12]:    '{S[0:12]}'")    # первые 12 символов
print(f"S[13:]:     '{S[13:]}'")     # с 13-го до конца
print(f"S[::2]:     '{S[::2]}'")     # каждый второй символ
print(f"S[::-1]:    '{S[::-1]}'")    # строка в обратном порядке
print(f"S[-6:]:     '{S[-6:]}'")     # последние 6 символов
````

### Глава 7. Списки и словари

**Вопрос:** В чём ключевое отличие списка от словаря?

**Ответ:** Список — упорядоченная коллекция элементов с доступом по числовому индексу (позиции). Словарь — коллекция пар «ключ-значение» с доступом по произвольному неизменяемому ключу. Списки оптимальны для упорядоченных наборов данных, словари — для ассоциативных данных, где важно быстрое обращение по ключу.


````python
# Список: доступ по позиции
colors = ['красный', 'зелёный', 'синий']
print(f"Список: colors[1] = '{colors[1]}'")

# Словарь: доступ по ключу
rgb = {'красный': '#FF0000', 'зелёный': '#00FF00', 'синий': '#0000FF'}
print(f"Словарь: rgb['зелёный'] = '{rgb['зелёный']}'")

# Словарь позволяет быстрый поиск O(1), список — O(n) для поиска по значению
print(f"'зелёный' in rgb: {('зелёный' in rgb)}")   # O(1)
print(f"'зелёный' in colors: {('зелёный' in colors)}")  # O(n)
````

### Глава 8. Кортежи и файлы

**Вопрос:** Зачем нужны кортежи, если есть списки?

**Ответ:** Кортежи неизменяемы, что даёт несколько преимуществ: они могут использоваться как ключи словаря и элементы множества; обеспечивают целостность данных (защиту от случайного изменения); занимают меньше памяти и работают быстрее списков. Кортежи уместны для фиксированных наборов данных, таких как координаты или записи из базы данных.


````python
import sys

# Кортеж как ключ словаря (список не может быть ключом!)
locations = {
    (53.9, 27.57): 'Минск',
    (55.75, 37.62): 'Москва'
}
print(f"Координаты Минска: {locations[(53.9, 27.57)]}")

# Сравнение размера в памяти
lst = [1, 2, 3, 4, 5]
tpl = (1, 2, 3, 4, 5)
print(f"\nРазмер списка [1..5]: {sys.getsizeof(lst)} байт")
print(f"Размер кортежа (1..5): {sys.getsizeof(tpl)} байт")
# Кортеж обычно занимает меньше памяти
````

### Глава 9. Общий обзор объектов

**Вопрос:** Что такое полиморфизм в контексте операций Python?

**Ответ:** Полиморфизм означает, что одна и та же операция (например, `+`, `len()`, итерация) может применяться к объектам разных типов, и результат зависит от типа операнда. Это позволяет писать универсальный код, который работает с разными типами данных.


````python
# Оператор + — полиморфный: конкатенация для строк/списков, сложение для чисел
print(f"3 + 4       = {3 + 4}")                    # сложение
print(f"'ab' + 'cd' = {'ab' + 'cd'}")              # конкатенация строк
print(f"[1,2] + [3] = {[1,2] + [3]}")              # конкатенация списков

# Функция len() — работает для любых коллекций
print(f"\nlen('hello')    = {len('hello')}")
print(f"len([1,2,3])    = {len([1,2,3])}")
print(f"len({{'a':1}})    = {len({'a':1})}")

# Оператор * — повторение для последовательностей, умножение для чисел
print(f"\n'ab' * 3    = {'ab' * 3}")
print(f"[0] * 5     = {[0] * 5}")
print(f"3 * 4       = {3 * 4}")
````
